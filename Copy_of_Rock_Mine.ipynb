**Objective :** Building a system in Python that can predict whether an object is either Rock or Mine with SONAR Data by using Logistic Regression Model**.**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #Fixed typo
```

```python
#loading data
sonar_data = pd.read_csv('/content/Copy of sonar data.csv', header= None)
```

```python
sonar_data.head() #this will display the starting 5 elements of
```

```python
sonar_data.shape #(row,col)
```

```python
sonar_data.describe() #describe--> statistical data #count is the no. elements in a row
```

```python
#to get the count of rocks and mines which are mentioned in the 60th colum.
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean() #this will group the 60th col. for the rock = r and mine = m
```

the difference between the rock and the mine mean are important difference to pridict wheater it's a r or m



```python
#now to spearte the lables and the data
X = sonar_data.drop(columns = 60 , axis =1 )
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

Training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape ,X_train.shape ,X_test.shape)
```

Model training --- logistic regression

```python
print(X_train ,Y_train ) #x_train is the training data and the y_train is the training label
```

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with the training data
model.fit(X_train , Y_train)
```

Checking the accuracy of the trained model

```python
#for accuracy
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print('The accurancy of the training data is : ' , training_data_accuracy) #accuracy >=70% counts good
```

Checking the accuracy of the test data

```python
#for accuracy
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
print('The accurancy of the testing data is : ' , testing_data_accuracy)  #accuracy >=70% counts good
```
